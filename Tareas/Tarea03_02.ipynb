```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# Punto 2

```python
class Lectura_csv:
    def __init__(self, limit,archivo):
        self.limit = limit
        self.val = 0
        self.archivo = archivo
    
    # Hace esta clase iterable
    def __iter__(self):
        return self
    
    # Hace esta clase un iterador
    def __next__(self):
        if self.val > self.limit:
            raise StopIteration
        else:
            return_val = self.val
            self.val += 2
            return return_val
    def lectura(self):
        archivo = self.archivo
        df = pd.read_csv(archivo,sep=',')
        df = pd.DataFrame(df)
        for col in df.columns:
            print(col)
    def stats(self):
        archivo = self.archivo
        df = pd.read_csv(archivo,sep=',')
        df = pd.DataFrame(df)
        q25 = []
        q75 = []
        #q25
        q25.append(np.percentile(df['sepal.length'],25))
        q25.append(np.percentile(df['sepal.width'],25))
        q25.append(np.percentile(df['petal.length'],25))
        q25.append(np.percentile(df['petal.length'],25))
        #-q75
        q75.append(np.percentile(df['sepal.length'],75))
        q75.append(np.percentile(df['sepal.width'],75))
        q75.append(np.percentile(df['petal.length'],75))
        q75.append(np.percentile(df['petal.length'],75))
        columnas = []
        for col in df.columns:
            columnas.append(col)
        mediana = []
        mediana.append(df['sepal.length'].mean())
        mediana.append(df['sepal.width'].mean())
        mediana.append(df['petal.length'].mean())
        mediana.append(df['petal.length'].mean())
        
        maximo = []
        maximo.append(df['sepal.length'].max())
        maximo.append(df['sepal.width'].max())
        maximo.append(df['petal.length'].max())
        maximo.append(df['petal.length'].max())
        
        minimo = []
        minimo.append(df['sepal.length'].min())
        minimo.append(df['sepal.width'].min())
        minimo.append(df['petal.length'].min())
        minimo.append(df['petal.length'].min())
        
        titles = ['sepal.length','sepal.width','petal.length','petal.width']
        for i in range(0,4):
            print(titles[i],'\nMediana = ',"%.2f"%mediana[i],'\nMaximo = ',maximo[i],'\nMinimo = ',minimo[i],
                 '\nPercentil 25 = ',"%.2f"%q25[i],'\nPercentil 75 = ',"%.2f"%q75[i],
                  '\n----------------------------------------------------------')
        val0 = []
        val1 = []
        val2 = []
        val3 = []
        val0.append(mediana[0])
        val0.append(maximo[0])
        val0.append(minimo[0])
        val0.append(q75[0])
        #1
        val1.append(mediana[1])
        val1.append(maximo[1])
        val1.append(minimo[1])
        val1.append(q75[1])
        #2
        val2.append(mediana[2])
        val2.append(maximo[2])
        val2.append(minimo[2])
        val2.append(q75[2])
        #3
        val3.append(mediana[3])
        val3.append(maximo[3])
        val3.append(minimo[3])
        val3.append(q75[3])
        fig,(ax0,ax1,ax2,ax3) = plt.subplots(nrows=4, ncols=1, figsize=(15, 25))
        ax0.hist(val0,color='orange')
        ax0.set_title('Sepal.length')
        ax1.hist(val1,color='blue')
        ax1.set_title('Sepal.width')
        ax2.hist(val2,color='green')
        ax2.set_title('Petal.length')
        ax3.hist(val3,color='black')
        ax3.set_title('Petal.width')
```

```python
archivo = Lectura_csv(16,'iris.csv')
archivo.lectura()
archivo.stats()
```
