```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
class Lectura_csv():
    def __init__(self, limit):
        self.limit = limit
        self.val = 0
    
    # Hace esta clase iterable
    def __iter__(self):
        return self
    
    # Hace esta clase un iterador
    def __next__(self):
        if self.val > self.limit:
            raise StopIteration
        else:
            return_val = self.val
            self.val += 2
            return return_val

```

```python
archivo = pd.read_csv('iris.csv')
print('INICIO DE LA LECTURA')
for i in Lectura_csv(16):
    try:
        display(pd.DataFrame(archivo.iloc[i]))
    except:
        None
print('\n NO HAY MÁS LÍNEAS PARES POR LEER')
```
